```python
import pandas as pd
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('data.csv',error_bad_lines=False)
data.head()
```

```python
data['strength'].unique()
```

```python
data.isna().sum()
```

```python
data[data['password'].isnull()]
```

```python
data.dropna(inplace=True)
data.isnull().sum()
```

```python
sns.countplot(x = data['strength'])
```

```python
password_tuple = np.array(data)
```

```python
password_tuple
```

```python
import random 
random.shuffle(password_tuple)
```

```python
x= [labels[0] for labels in password_tuple]
y= [labels[1] for labels in password_tuple]
```

```python
x
```

```python
def word_divide_char(inputs):
    character = []
    for i in inputs:
        character.append(i)
    return character
```

```python
word_divide_char('kzde5577')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(tokenizer = word_divide_char)
X = vectorizer.fit_transform(x)
```

```python
X.shape
```

```python
X[2]
```

```python
vectorizer.get_feature_names_out()
```

```python
first_document_vector = X[0]
first_document_vector
```

```python
first_document_vector.T.todense()
```

```python
df = pd.DataFrame(first_document_vector.T.todense(), index =vectorizer.get_feature_names_out(), columns = ['TF-IDF'] )
df.sort_values(by = 'TF-IDF', ascending = False, inplace = True)
df 
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
x_train.shape # 80%
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=0, multi_class='multinomial').fit(x_train, y_train)
clf 
```

```python
dt = np.array(['%@123abcd'])
pred = vectorizer.transform(dt)
clf.predict(pred)
```

```python
y_pred = clf.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(accuracy_score(y_test, y_pred)) 
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
